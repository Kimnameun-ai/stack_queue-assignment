1. LeetCode 225. Implement Stack using Queues

```python
from collections import deque

class MyStack:

    def __init__(self):
        self.q = deque()

    def push(self, x: int) -> None:
        self.q.append(x)
        # 새로 들어온 원소를 맨 앞으로 이동시키기
        for _ in range(len(self.q) - 1):
            self.q.append(self.q.popleft())

    def pop(self) -> int:
        return self.q.popleft()

    def top(self) -> int:
        return self.q[0]

    def empty(self) -> bool:
        return not self.q
```

2. LeetCode 232. Implement Queue using Stacks

```python
class MyQueue:

    def __init__(self):
        self.stack_in = []
        self.stack_out = []

    def push(self, x: int) -> None:
        self.stack_in.append(x)

    def pop(self) -> int:
        self.peek()  # stack_out에 데이터가 없으면 옮기기
        return self.stack_out.pop()

    def peek(self) -> int:
        if not self.stack_out:
            while self.stack_in:
                self.stack_out.append(self.stack_in.pop())
        return self.stack_out[-1]

    def empty(self) -> bool:
        return not self.stack_in and not self.stack_out
```

```python
# ============================================
# 연습문제 01: 리스트의 뒤를 front, 앞을 tail로 간주한 큐 구현
# ============================================

class ListQueue:
    def __init__(self):
        self._queue = []

    def enqueue(self, x):
        # tail 방향으로 추가 → 리스트의 앞에 삽입
        self._queue.insert(0, x) 

    def dequeue(self):
        # front 방향 → 리스트의 뒤에서 꺼냄
        return self._queue.pop()  # O(1)

    def front(self):
        # front는 리스트의 맨 뒤
        return self._queue[-1]

    def isEmpty(self) -> bool:
        return len(self._queue) == 0

    def dequeueAll(self):
        self._queue.clear()


# ============================================
# 연습문제 02: 큐를 이용한 집합 포함 여부 검사
# ============================================
from collections import deque

def is_in_set(input_str, w$w):
    queue = deque()  # 1. 비어있는 큐 생성

    for ch in input_str:
        queue.append(ch)  # 2. input_str의 각 문자를 큐에 삽입

    while queue:  # 3. 큐가 비어 있지 않을 동안 반복
        ch = queue.popleft()  # 4. 큐의 앞(front)에서 문자 하나 꺼냄

        if ch not in w$w:     # 5. 꺼낸 문자가 w$w 집합에 없으면
            return False      #    → 포함되지 않은 것이므로 False 반환

    return True  # 6. 반복이 끝날 때까지 모두 포함되었다면 True 반환

# 예시:
# print(is_in_set("wow", {"w", "s", "o"}))  # True


# ============================================
# 연습문제 03: 큐 복사하기 (깊은 복사)
# ============================================
def copy_queue(a):
    from collections import deque
    b = deque()
    for item in a:
        b.append(item)
    return b

# 예시:
# q1 = deque([1,2,3])
# q2 = copy_queue(q1)
# print(q2)  # deque([1, 2, 3])


# ============================================
# 연습문제 04: 2개의 큐로 스택 구현
# ============================================
class StackWithQueues:
    def __init__(self):
        self.q1 = deque()
        self.q2 = deque()

    def push(self, x):
        self.q1.append(x)

    def pop(self):
        while len(self.q1) > 1: #q1에 요소가 2개 이상이면, 마지막 요소 빼고 q2로 옮기기 시작
            self.q2.append(self.q1.popleft()) #q1의 앞에서 하나 빼서 q2 뒤에 넣음 (FIFO 방식)
        result = self.q1.popleft() #이제 q1에는 마지막 요소 하나만 남음 → 이걸 꺼냄 (= 스택 pop과 동일)
        self.q1, self.q2 = self.q2, self.q1 #q2가 다시 주 큐(q1)가 되고, q1은 보조 큐(q2)로 바꿈
        return result #마지막에 들어온 값을 반환함 (stack의 pop 결과)

# 예시:
# s = StackWithQueues()
# s.push(1); s.push(2)
# print(s.pop())  # 2


# ============================================
# 연습문제 05: 2개의 스택으로 큐 구현
# ============================================
class QueueWithStacks:
    def __init__(self):
        self.s1 = []
        self.s2 = []

    def enqueue(self, x):
        self.s1.append(x)

    def dequeue(self):
        if not self.s2:
            while self.s1:
                self.s2.append(self.s1.pop())
        return self.s2.pop()

# 예시:
# q = QueueWithStacks()
# q.enqueue(1); q.enqueue(2)
# print(q.dequeue())  # 1


# ============================================
# 연습문제 06: 효율적인 Deque 구현
# ============================================

'''답: 코드 [7-13]은 원형 연결 리스트를 이용해서 큐를 구현한 클래스이다. 
Dequeue의 경우 앞과 뒤 양쪽에서 삽입, 삭제가 가능하므로 dequeue의 앞에 삽입할때는 O(1),'
뒤에 삽입할 때는 O(1), 앞에서 삭제할 때는 O(1), 뒤에서 삭제할 때는 마지막 노드까지 탐색해야 하므로 O(n) 
수행시간이 소요된다.
'''


# ============================================
# 연습문제 07: LinkedListBasic을 사용한 deque 구현
# ============================================
'''[코드 5-15]의 LinkedListBasic 클래스는 단방향 단일 연결 리스트이고,
노드를 앞에서부터 순차적으로 탐색해야 한다. 
큐의 큐기를 n이라고 할 때, enqueue(): 맨 뒤 삽입의 경우 
맨 끝 노드를 찾아야 하므로 시간 복잡도는 O(n),
dequeue(): 맨 앞 삭제의 경우 
head=head.next하면 되므로 시간 복잡도는 O(1)이다.
'''


# ============================================
# 연습문제 08: ListQueue 클래스 완성
# ============================================
class Deque:
    def __init__(self):
        self._deque = []

    def isEmpty(self) -> bool:
        return len(self._deque) == 0

    def enqueueRear(self, x):
        """뒤쪽에 삽입"""
        self._deque.append(x)

    def enqueueFront(self, x):
        """앞쪽에 삽입"""
        self._deque.insert(0, x)

    def dequeueFront(self):
        """앞쪽에서 삭제"""
        if self.isEmpty():
            return None
        return self._deque.pop(0)

    def dequeueRear(self):
        """뒤쪽에서 삭제"""
        if self.isEmpty():
            return None
        return self._deque.pop()

    def front(self):
        """맨 앞 요소 확인"""
        if self.isEmpty():
            return None
        return self._deque[0]

    def rear(self):
        """맨 뒤 요소 확인"""
        if self.isEmpty():
            return None
        return self._deque[-1]

    def dequeueAll(self):
        """모든 요소 제거"""
        self._deque.clear()

    def printDeque(self):
        print("Deque from front: ", end='')
        for item in self._deque:
            print(item, end=' ')
        print()

```
